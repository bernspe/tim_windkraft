```python
def einlesen_exps(exp_verz):
    alleDat=list(filter(lambda x: x.endswith('.csv'), os.listdir(exp_verz)))
    suparray=np.zeros((len(alleDat),2000,4))
    if (len(alleDat)==0):
        return np.nan
    for i in range(len(alleDat)):
        a=np.genfromtxt(exp_verz+'/'+alleDat[i],delimiter=',')
        a*=volt_factor
        suparray[i,:,:]=a
    return suparray

exp_df['Arrays']=exp_df['Verzeichnis'].apply(lambda x: einlesen_exps(x))
```

```python
def schnelllaufzahl(x):
    umdrehungen=[]
    for i in range (x.shape[0]):
        a=x[i]
        wr=np.where(np.diff(np.where((a[:,1])<2000,0,1))>0,0,1)
        umdr_wr=(x.shape[1]-np.count_nonzero(wr))
        umdrehungen.append(umdr_wr)
    return np.mean(np.array(umdrehungen))

def schnelllaufzahl_std(x):
    umdrehungen=[]
    for i in range (x.shape[0]):
        a=x[i]
        wr=np.where(np.diff(np.where((a[:,1])<2000,0,1))>0,0,1)
        umdr_wr=(x.shape[1]-np.count_nonzero(wr))
        umdrehungen.append(umdr_wr)
    return np.array(umdrehungen)

def anemometer1(x):
    umdrehungen=[]
    for i in range (x.shape[0]):
        a=x[i]
        a_1=np.where(np.diff(np.where((a[:,2])<500,0,1))>0,1,0)
        umdr_wr=(np.count_nonzero(a_1))
        umdrehungen.append(umdr_wr)
    return np.array(umdrehungen)

def anemometer2(x):
    umdrehungen=[]
    for i in range (x.shape[0]):
        a=x[i]
        a_2=np.where(np.diff(np.where((a[:,3])<500,0,1))>0,1,0)
        umdr_wr=(np.count_nonzero(a_2))
        umdrehungen.append(umdr_wr)
    return np.array(umdrehungen)

def wirkungsgrad(x):
    umdrehungen=[]
    0.5*(1-(exp_df[anemometer2]**2/exp_df[anemometer1]**2))*(1+(exp_df[anemometer2]**2/exp_df[anemometer1]**2))

#exp_df['Mean_LeistungWindrad (in mW)']=(exp_df['Mean_SpannungWindrad (in mV)']**2)/10000
#exp_df['Standardabweichung_Leistung']=

exp_df['Mean_SchnelllaufzahlWindrad']=exp_df['Arrays'].apply(schnelllaufzahl)/exp_df['AnzahlRotorblaetter']
exp_df['Standardabweichung_Schnelllaufzahl']=exp_df['Arrays'].apply(schnelllaufzahl_std)

exp_df['Mean_Anemometer1']=exp_df['Arrays'].apply(np.mean)#(anemometer1)
exp_df['Standardabweichung_Anemometer1']=exp_df['Arrays'].apply(np.std)#(anemometer1)

exp_df['Mean_Anemometer2']=exp_df['Arrays'].apply(np.mean)#(anemometer2)
exp_df['Standardabweichung_Anemometer2']=exp_df['Arrays'].apply(np.std)#(anemometer2)

exp_df['Wirkungsgrad']=exp_df['Arrays'].apply(np.mean)#(wirkungsgrad)
exp_df['Standardabweichung_Wirkungsgrad']=exp_df['Arrays'].apply(np.std)#(wirkungsgrad)
```
